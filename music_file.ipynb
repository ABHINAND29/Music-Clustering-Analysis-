# Features Selection
## Goal: Select features that represent the sound of a song and are useful for clustering.
### Recommended features:danceability,energy,loudness,speechiness,acousticness,instrumentalness,liveness,valence,tempo,duration_ms


```python
features = [
    'danceability', 'energy', 'loudness', 'speechiness',
    'acousticness', 'instrumentalness', 'liveness',
    'valence', 'tempo', 'duration_ms'
]

X = df[features]
```

```python
df[features].dtypes
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Dimensionality Reduction (Optional but Recommended for Visualization)
### Goal: Reduce the dataset to 2 or 3 components to visualize clusters later
#### PCA (Principal Component Analysis):

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

##  K-Means Clustering

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 10), wcss)
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X_scaled)

df['cluster'] = labels
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_scaled, labels)  # or X_reduced if using PCA
print("Silhouette Score:", score)
```

```python

sns.scatterplot(x=df['tempo'], y=df['energy'], hue=df['cluster'])
plt.show()
```

🎧 Music Clustering Analysis – Summary Report
📌 Objective

The goal of this project was to group songs into meaningful clusters based on their audio features such as danceability, energy, tempo, acousticness, and valence. This helps in understanding different types of music and enables applications like recommendation systems.

⚙️ Methodology
Feature Selection
Selected relevant audio features:
danceability, energy, loudness, speechiness
acousticness, instrumentalness, liveness
valence, tempo
Data Preprocessing
Removed non-relevant columns (track name, artist, IDs)
Scaled features using StandardScaler
Clustering Algorithm
Applied K-Means Clustering
Optimal number of clusters (k = 3) determined using:
Elbow Method
Silhouette Score
Evaluation
Silhouette Score used to measure cluster quality
PCA/t-SNE used for visualization
📊 Cluster Overview
🔸 Cluster 0: High-Energy Party Tracks

Characteristics:

High energy and danceability
Higher tempo
Moderate to high loudness
Positive valence

Interpretation:
These are upbeat, energetic songs suitable for parties, workouts, and clubs.

🔸 Cluster 1: Chill Acoustic / Calm Tracks

Characteristics:

High acousticness
Low energy
Lower tempo
Softer loudness

Interpretation:
These songs are calm, relaxing, and often acoustic. Ideal for studying, relaxing, or background listening.

🔸 Cluster 2: Happy / Balanced Tracks

Characteristics:

Moderate energy
High valence (positive mood)
Balanced danceability
Medium tempo



Interpretation:
These are feel-good songs with a balanced mix of rhythm and positivity, suitable for everyday listening.

📈 Visualization Insights
PCA / t-SNE plots showed reasonable separation between clusters
Heatmaps highlighted clear differences in energy, acousticness, and tempo
Distribution plots revealed:
Cluster 0 → tightly packed high-energy songs
Cluster 1 → low-energy, high-acoustic spread
Cluster 2 → moderate and balanced distributions
📉 Model Evaluation
Silhouette Score: Indicates moderate clustering quality
Some overlap exists between clusters (expected in music data)
Overall clusters are interpretable and meaningful
🎯 Key Insights
Songs can be effectively grouped based on audio features
Energy, acousticness, and tempo are the strongest differentiators
Clusters align well with real-world music categories:
Party
Chill
Happy
🚀 Applications

📊 Music Trend Analysis
🎶 Genre or mood classification

Some clusters overlap due to similarity in songs

Results depend on selected features
✅ Conclusion

The clustering approach successfully grouped songs into meaningful categories based on their audio characteristics. Despite minor overlaps, the clusters are interpretable and useful for real-world applications such as recommendation systems and playlist generation.
